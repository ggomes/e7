```python
import numpy as np
```

NumPy offers two equivalent approaches.
+ `@` operator
+ `np.linalg.matmul`

### Example: Matrix times column vector

\begin{equation*}
A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}
\end{equation*}

\begin{equation*}
 x_{col} = \begin{bmatrix} 2 \\ -1 \\ 0\end{bmatrix} 
\end{equation*}

\begin{equation*}
A\:x_{col} = \begin{bmatrix} 0 \\ 3 \end{bmatrix} 
\end{equation*}

```python
A = np.array([[1,2,3],[4,5,6]])

x1D = np.array([2,-1,0])
xcol = np.array([[2],[-1],[0]])

print("\nwith xcol and @:")
print(A@xcol)

print("\nwith xcol and matmul: ")
print(np.linalg.matmul(A,xcol))

print("\nwith x1D and @: ")
print(A@x1D)
print("\nwith x1D and matmul: ")
print(np.linalg.matmul(A,x1D))
```

### Example: Row vector times matrix

\begin{equation*}
A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}
\end{equation*}

\begin{equation*}
\hspace{1em} x_{row} = \begin{bmatrix} 2 & -1 \end{bmatrix} 
\end{equation*}

\begin{equation*}
x_{row}A = \begin{bmatrix} -2 & -1 & 0 \end{bmatrix} 
\end{equation*}

```python
A = np.array([[1,2,3],[4,5,6]])

x1D = np.array([2,-1])
xrow = np.array([[2,-1]])

print("\nwith xrow and @:")
print(xrow@A)

print("\nwith xrow and matmul:")
print(np.linalg.matmul(xrow,A))

print("\nwith x1D and @:")
print(x1D@A)
print("\nwith x1D and matul:")
print(np.linalg.matmul(x1D,A))
```
