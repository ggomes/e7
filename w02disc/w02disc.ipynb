# 1. Number adding program

### Version 1

```python
print("Welcome to the number adding program!")
total = 0
add_string = input("Enter a number to add (or type 'stop'): ") 
while add_string!="stop":
    add = float(add_string)
    total += add
    print("Total: ",total)
    add_string = input("Enter a number to add (or type 'stop'): ") 
    print(add_string)
```

# 3. More about functions

## 3.a) Multiple outputs

Return multiple values by comma-separating them.

```python
def myFunction1(a,b):
    x = a+b
    y = a*b
    z = a%b
    return x, y, z
```

```python
r,s,t = myFunction1(10,3)
```

The return keyword can appear anywhere in the body of the function.

```python
def myFunction2(x):
    if x<1:
        return x
    else:
        return x**2
```

## 3.b) Default input values 

You can specify default values for inputs in the **def** line of the function.

```python
def myFunction3(a,b=1,c=0):
    return a*b + c
```

## 3.c) Positional vs. keyword styles for passing inputs

```python
def myFunction4(a,b,c):
    print("a =",a)
    print("b =",b)
    print("c =",c)
```
