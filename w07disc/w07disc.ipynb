```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from matplotlib import cm
```

## 1. Adding legends to your plots

+ Must pass `label` argument to each graphical element in the legend.
+ Here we are also adjusting the spines by
    + turning off the top and right spines
    + setting the position of the bottom and left spines to `zero`.

```python
x = np.linspace(0,8,100)
y1 = np.exp(-x)*np.sin(x)
y2 = np.exp(-0.8*x)*np.sin(1.5*x)

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(x,y1,linewidth=2,color='red',linestyle=':')
ax.plot(x,y2,linewidth=2,color='blue',linestyle='--')

ax.set_ylabel('displacement [cm]',fontsize=12)
ax.set_xlabel('time [sec]',loc='right',fontsize=12)
ax.spines[['top','right']].set_visible(False)
ax.spines[['bottom','left']].set_position('zero')
```

## 2. Animating the flight of a drone 

### a) Load the data

```python
traj = np.load('drone_trajectory.npy')
time = traj[:,0]
x = traj[:,1]
y = traj[:,2]
z = traj[:,3]
```

### b) Initialize the figure

+ `ax.plot()` returns a tuple of length 2, and with the line object in the first position. 
+ Setters on the line object include 
    + `set_data_3d`
    + `set_color`
    + `set_linestyle`
    + `set_linewidth`
    + `set_marker`
    + `set_markeredgecolor`
    + `set_markeredgewidth`
    + `set_markerfacecolor`
    + `set_markerfacecoloralt`
    + `set_markersize`
    + `set_alpha`
    + etc.


```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(projection='3d')
line, = ax.plot(x, y, z, c='black',linewidth=1, linestyle='-')
point, = ax.plot(x[0], y[0], z[0], color='green', marker='o')
```

### c) Figure update function

```python
def update(frame):
    line.set_data_3d(x[:frame+1],y[:frame+1],z[:frame+1])
    point.set_data_3d([x[frame]],[y[frame]],[z[frame]])   
    if frame%2==0:
        point.set_color('red')
    else:
        point.set_color('blue')
```

### d) Create the animation

```python
movie = FuncAnimation(fig=fig, 
                      func=update, 
                      frames=len(x), 
                      interval=100 )
```

### e) Display the animation

```python
HTML(movie.to_html5_video())
```

### e) Save the animation

```python
movie.save('drone_movie.gif')
```

### c) Surface plot

+ `ax.plot_surface` returns the surface object.
+ It is difficult to dynamically change the data in a surface plot. 
+ It is always possible to clear and remake the entire plot in the update function. This is slower but still valid.  
+ Use `ax.clear()` to clear the axes. 

```python
# data ...............................................
x = np.arange(-3,3,0.25)
y = np.arange(-3,3,0.25)
X, Y = np.meshgrid(x, y)
Z = np.cos(np.sqrt(X**2 + Y**2)) + 1

# initialize the figure ..............................
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.RdYlBu,vmin=-1, vmax=1)
ax.set_zlim(-1,1)
```
