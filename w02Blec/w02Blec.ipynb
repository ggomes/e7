# Speeding ticket

### Version 1

```python
speed_limit = 65
speed = 68
has_license = False

if speed>speed_limit:
    if has_license:
        print("You get a ticket!")
    else:
        print("You are under arrest!")
else:
    if speed<0:
        print("I have a few questions for you...")
    else:
        print("Have a nice day!")
```

# Number adding program

### Version 1

```python
print("Welcome to the number adding program!")
total = 0
add_string = input("Enter a number to add (or type 'stop'): ") 
while add_string!="stop":
    add = float(add_string)
    total += add
    print("Total: ",total)
    add_string = input("Enter a number to add (or type 'stop'): ") 
    print(add_string)
```

### Version 2

```python
print("Welcome to the number adding program!")
total = 0
while True:
    add_string = input("Enter a number to add (or type 'stop'): ") 
    if add_string=='stop':
        break
    add = float(add_string)
    total += add
    print("Total: ",total)
```
