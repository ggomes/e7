```python
import scipy
import matplotlib.pyplot as plt
```

## Sampling a normal distribution 

```python
# Create the random variable with scipy.stats.norm
X = scipy.stats.norm( 
    loc=2,         # mean
    scale=1)       # standard deviation

# sample it with "rvs"
sample = X.rvs(size=100)

# plot
_, ax = plt.subplots(figsize=(6,1))
ax.plot(sample,0*sample,'+')
ax.spines[:].set_visible(False)
ax.set_yticks([])
```

## Histogram

```python
_, ax = plt.subplots(figsize=(5,2))
ax.hist(sample,rwidth=0.8)
ax.spines[['top','right']].set_visible(False)
```

## Sampling a unifom distribution 

```python
# Create the random variable with scipy.stats.uniform
X = scipy.stats.uniform( 
    loc=2,         # mean
    scale=1)       # standard deviation

# sample with "rvs"
sample = X.rvs(size=100)

# plot
_, ax = plt.subplots(figsize=(6,1))
ax.plot(sample,0*sample,'+')
ax.spines[:].set_visible(False)
ax.set_yticks([])

# histogram
_, ax = plt.subplots(figsize=(5,2))
ax.hist(sample,rwidth=0.8)
ax.spines[['top','right']].set_visible(False)
```

## Sampling a general discrete distribution 

```python
# Create the random variable with scipy.stats.rv_discrete
X = scipy.stats.rv_discrete(values=([1,2,3], [0.3,0.3,0.4]))

# sample with "rvs"
x = X.rvs(size=100)      # number of samples

# histogram
_, ax = plt.subplots(figsize=(4,2))
ax.hist(x,rwidth=1.0,align='left')
ax.set_xticks([1,2,3])
ax.set_xlim(0,4)
ax.spines[['top','right']].set_visible(False)
```
