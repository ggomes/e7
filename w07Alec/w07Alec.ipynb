```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from pathlib import Path

# load some data
with open(Path.cwd()/'plotdata.pickle','rb') as f:
    alldata = pickle.load(f)

x_line = alldata['x_line']
y_line1 = alldata['y_line1']
y_line2 = alldata['y_line2']
x_scatter = alldata['x_scatter']
y_scatter = alldata['y_scatter']
x_bar = alldata['x_bar']
y_bar = alldata['y_bar']
data_hist = alldata['data_hist']
```

## 1) Three methods for creating a figure and axes

```python
x = np.linspace(0,100,200)
y1 = np.sin(x/10)
y2 = np.exp(-x/50)*np.cos(x)
```

#### 1.1) `plt.figure()` + `plt.subplot()` 

Pros:
+ Matlab-like
+ Plots can span multiple grid cells

Cons:
+ State-full

```python
plt.figure(figsize=(6,4))
plt.subplot(2,1,1)
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)
```

#### 1.2) `plt.subplots()`

Pros:
+ State-less
+ Share x-axis and y-axis tick labels with `sharex` and `sharey`

Cons:
+ Grids layouts only with `nrows` and `ncols`.

**Note**
+ This is the recomended method for most situations. 
+ You can use `GridSpec` to get find control over axes locations (not covered in E7).

```python
fig, axs = plt.subplots(figsize=(6,4),nrows=2,ncols=1,sharex=True)
axs[0].plot(x,y1)
axs[1].plot(x,y2)
```

#### 1.3) `fig = plt.figure()` + `fig.add_subplot()`

Pros:
+ State-less
+ Can construct general layouts

Cons:
+ Grids layouts only with `nrows` and `ncols`.
+ No sharing x-axis and y-axis tick labels. 

```python
fig = plt.figure(figsize=(6,4))
ax1 = fig.add_subplot(2,1,1)
ax1.plot(x,y1)

ax2 = fig.add_subplot(2,1,2)
ax2.plot(x,y2)
```

## Four basic types of graphical elements

+ [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
+ [bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
+ [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)
+ [hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) 



```python
fig, axs = plt.subplots(figsize=(6,6),nrows=2,ncols=2)
axs[0,0].plot(x_line,y_line1)
axs[0,0].plot(x_line,y_line2)
axs[1,0].scatter(x_scatter,y_scatter)
axs[0,1].bar(x_bar,y_bar)
axs[1,1].hist(data_hist)
```

## Three additional types of grpahical elements

+ [imread](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imread.html)
+ [stem](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stem.html)
+ [fill_between](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html)

```python
image = plt.imread(Path.cwd()/'campanile.jpg')

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(2*x)
```

```python
fig, axs = plt.subplots(figsize=(8,3),nrows=1,ncols=3)

axs[0].imshow(image)
axs[1].stem(x[::4],y1[::4])
axs[2].fill_between(x, y1,y2,alpha=0.4)
```
