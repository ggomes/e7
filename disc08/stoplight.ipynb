```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
```

# 1. Setup

```python
# parameters ..................
num_cells = 6
num_timesteps = 10

# allocate memory .............
s = np.zeros((num_timesteps,num_cells))
```

# 2. Initial condition

```python
s[0,:] = np.array([2,0,1,1,0,2])
```

# 3. Run the simulation 

```python
for k in range(0,num_timesteps-1):

    for i in range(1,num_cells-1):

        if s[k,i]==0:
            if s[k,i-1]==2:
                s[k+1,i] = 2
            else:
                s[k+1,i] = 0

        if s[k,i]==1:
            if s[k,i+1]==0:
                s[k+1,i] = 2
            else:
                s[k+1,i] = 1

        if s[k,i]==2:
            s[k+1,i] = s[k,i+1]
```

# 4. Plot

```python
cmap = colors.ListedColormap(['white','red','green'])

fig, ax = plt.subplots(figsize=(12,10))
ax.imshow(s,cmap=cmap)
ax.set_xticks(range(num_cells))
ax.set_yticks(range(num_timesteps))
ax.set_xlabel('cell',fontsize=20)
ax.set_ylabel('time step',fontsize=20)
```
