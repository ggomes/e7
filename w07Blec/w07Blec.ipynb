### Example

Create a surface plot for the following function with domain $x\in[-3,3]$ and $y\in[-3,3]$

$$ f(x,y) = \cos\left( \sqrt{x^2+y^2}\right) $$

```python
from matplotlib import cm

x = np.arange(-3,3,0.25)
y = np.arange(-3,3,0.25)
X, Y = np.meshgrid(x, y)

Z = np.cos(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(5,10))

ax = fig.add_subplot(2,1,1,projection='3d')
ax.plot_wireframe(X, Y, Z)

ax = fig.add_subplot(2,1,2,projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
```

## Surface plot on an irregular grid

### Load some 3D data

```python
data = np.load(Path.cwd()/'data.npz')
x = data['x']
y = data['y']
z = data['z']
```

### Create the irregular XY mesh

```python
import matplotlib.tri as mtri

irr_mesh = mtri.Triangulation(x, y)
```

```python
fig, ax = plt.subplots(figsize=(3,3))
ax.triplot(irr_mesh, marker = '.', markeredgecolor = 'black');
```

### Plot the surface on the irregular grid

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')

ax.plot_trisurf(irr_mesh, z, cmap=cm.coolwarm, 
                           edgecolor='grey')

ax.scatter(x,y,z, marker='.', c='black')
```

## Animation

```python
import matplotlib.animation as animation
from IPython.display import HTML

# data ...............................................
xbar = np.arange(10)
ybar = np.sin(0.5*xbar)
xline = np.linspace(0,10)
yline = np.sin(0.5*xline)

# initialize the figure ..............................
fig, ax = plt.subplots(figsize=(5,5))
bar = ax.bar(xbar,ybar)
line, = ax.plot(xline,yline,linewidth=4,color='purple')

# figure update function .............................
def update(frame):
    yline = np.sin(0.5*xline - frame/20)
    line.set_ydata(yline)

    ybar = np.sin(0.5*xbar - frame/20)
    for val, rect in zip(ybar,bar.get_children()):
        rect.set_height(val)

# create the animation ...............................
movie = animation.FuncAnimation(fig=fig, 
                              func=update, 
                              frames=200, 
                              interval=20 )

# animation player ...................................
HTML(movie.to_html5_video())
```
