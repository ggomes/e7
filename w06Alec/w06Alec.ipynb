### Writing arbitrary text

```python
outfile = Path.cwd() / 'write_test.txt'
with open(outfile,'w') as f:
    f.write('First line\n')
    f.write('Second line\n')
```

### Reading arbitrary text

```python
infile = Path.cwd() / 'write_test.txt'
with open(infile,'r') as f:
    for line in f.readlines():
        print(line)
```

### The [pickle](https://docs.python.org/3/library/pickle.html) package

```python
import pickle
```

```python
X = [{ 'type' : 'A',
       'size' : 29,
       'cost' : 1.50 },
     { 'type' : 'B',
       'size' : 19,
       'cost' : 1.30 } ]
Y = 'This is a string.'
```

```python
outfile = Path.cwd()/'my_data.pickle'
with open(outfile,'wb') as f:
    pickle.dump((X,Y),f)
```

```python
infile = Path.cwd()/'my_data.pickle'
with open(infile,'rb') as f:
    X, Y = pickle.load(f)
```
