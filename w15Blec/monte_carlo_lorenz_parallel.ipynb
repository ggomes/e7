```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.stats import norm
import time
```

\begin{align*}
x' &= \sigma(y-x) \\ 
y' &= x(\rho-z)-y \\
z' &= xy - \beta z
\end{align*}

## Global constants

```python
# physical parameters
sigma=10
rho=28
beta=8/3

# simulation parameters
T = 30       # [s] total simulation time
dt = 0.15    # [s] time time step
t_eval = np.arange(0,T,dt)

# initial condition
x0 = -10
y0 = -10
z0 = 30
```

## State equaton $\mathbf{y}' = f(t,\mathbf{y})$

```python
def f(t, Y):
    x, y, z = Y
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return dxdt, dydt, dzdt
```

## Create random variables for initial condition 

```python
X0 = norm(loc=x0,scale=0.1)
Y0 = norm(loc=y0,scale=0.1)
Z0 = norm(loc=z0,scale=0.1)

# ensemble size
n = 50
```

# Serial Monte Carlo run

```python
start = time.perf_counter()

# allocate
Xens = np.empty((len(t_eval),n))
Yens = np.empty((len(t_eval),n))
Zens = np.empty((len(t_eval),n))

# iterate over the ensemble
for i in range(n):

    # IVP solver
    sol = solve_ivp(f, t_span=(0,T), 
                    y0=(X0.rvs(),Y0.rvs(),Z0.rvs()), 
                    t_eval=t_eval)

    # unpack and store the solution
    Xens[:,i] = sol.y[0,:]
    Yens[:,i] = sol.y[1,:]
    Zens[:,i] = sol.y[2,:]

end = time.perf_counter()
print(f"Execution time: {end - start:.6f} seconds")
```
