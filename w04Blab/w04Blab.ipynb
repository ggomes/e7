## Part 1.5 -- Standard deviations

Sensors s0 and s2 are supposedly kept at constant temperatures, but we can see from the figure that there is some fluctuation or *noise* in the measurements. We can quantify this noise in the signal by computing its standard deviation. The standard deviation $\sigma$ of an array $A=[a_1,...,a_N]$ is defined as:

$$ \sigma = \sqrt{ \frac{1}{N-1} \sum_{i=1}^{N} (a_i-\bar{a})^2 } $$

Here $\bar{a}$ is the mean of the elements in $A$, and $N$ is the number of elements in $A$.

## Part 2.4 -- Loop through time

We can now code the "for" loop that will compute the trajectory of the ball through time. In the `k`'th iteration, the task is to compute values to write into the `k`'th position of our six arrays, based on the values in the `k-1`'th positions.

The calculation is done in four steps:

1. **Rolling update**. The ball is considered to be rolling at time instant `k` if its position at `k-1` is within one centimeter of the floor *and* its vertical speed at `k-1` is less than 0.01 m/s in absolute value. 

2. **Position update**. We assume that `vx[k-1]` and `vy[k-1]` are the *average* velocity of the ball over the time interval from `time[k-1]` to `time[k]`. The ball therefore advances by `dt*vx[k-1]` in that interval, and hence `px[k] = px[k-1]+dt*vx[k-1]` (here `dt` is the time step of 0.015 second). The same is true for the vertical position `py[k]`, provided the ball is not rolling at time `k-1`. If it is rolling, then the vertical position at time `py[k]` is set to zero.

2. **Velocity update**. The horizontal speed of the ball remains constant for all time. The ball's change in vertical speed from `k-1` to `k` depends on whether it was rolling or not at time `k-1`. If it was rolling, then its vertical speed is zero. Otherwise its vertical speed is decreased by `dt*g` with respect to the previous time instant, where `g` is the acceleration of gravity, ie. $9.81 \; m/s^2$.


4. **Bounce detection**. A bounce occurs at time `k` whenever all three of the following conditions are met:
    + The vertical velocity at time `k` is less-or-equal than zero,
    + The vertical position at time `k` is less-or-equal than zero,
    + The ball is not rolling at time `k`.

5. **Bounce action**. If a bounce is detected at time `k`, then 
    + The vertical velocity is flipped from negative to positive, and multiplied by an energy-dissipating factor of 0.3.
    + The vertical position at time `k` should be set to 0.
