### Debgugging example

The function `find_max` is supposed to find the largest of the values in `numbers`. 

````python
def find_max(numbers):
    max_value = 0
    
    for n in numbers:
        if n > max_value:
            print(f"n={n} max_value={max_value}")
            max_value = n
   
    return max_value
````

````python
themax = find_max([5,3,7])

print( f"Maximum: {themax}" )
````
