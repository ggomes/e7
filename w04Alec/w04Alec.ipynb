## Broadcasting

$2D \bigotimes 1D$

```python
A = np.array([[0,0,0],
              [1,1,1]])

B = np.array([1,2,3])
```

$2D \bigotimes 2D$ row

```python
A = np.array([[0,0,0],
              [1,1,1]])

B = np.array([[1,2,3]])
```

$2D \bigotimes 2D$ columns

```python
A = np.array([[0,0,0],
              [1,1,1]])

B = np.array([[1],[2]])
```
